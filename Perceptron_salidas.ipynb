#Perceptr√≥n

## Ejemplo simple con AND

```python
# Paso 1: Importar librer√≠as
import numpy as np
import matplotlib.pyplot as plt

# Paso 2: Definir datos para la compuerta AND
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 0, 0, 1])  # Salidas para AND

# Paso 3: Funci√≥n de activaci√≥n (escal√≥n)
def step(x):
    return np.where(x >= 0, 1, 0)

# Paso 4: Definir clase Perceptr√≥n simple
class SimplePerceptron:
    def __init__(self, lr=0.1, epochs=10):
        self.lr = lr
        self.epochs = epochs
        self.weights = None

    def fit(self, X, y):
        X_bias = np.hstack((np.ones((X.shape[0], 1)), X))  # Agregar columna de 1's (sesgo)
        self.weights = np.zeros(X_bias.shape[1])
        for _ in range(self.epochs):
            for xi, target in zip(X_bias, y):
                prediction = step(np.dot(xi, self.weights))
                self.weights += self.lr * (target - prediction) * xi

    def predict(self, X):
        X_bias = np.hstack((np.ones((X.shape[0], 1)), X))
        return step(np.dot(X_bias, self.weights))

# Paso 5: Entrenar el modelo
model = SimplePerceptron()
model.fit(X, y)
y_pred = model.predict(X)

print("Predicciones:", y_pred)

# Paso 6: Graficar la frontera de decisi√≥n
def plot_decision_boundary(X, y, model):
    plt.figure(figsize=(6, 6))
    for i, point in enumerate(X):
        color = 'red' if y[i] == 1 else 'blue'
        plt.scatter(point[0], point[1], c=color, s=100, edgecolors='k')

    x_vals = np.array([-0.1, 1.1])
    w = model.weights
    if w[2] != 0:  # evitar divisi√≥n por cero
        y_vals = -(w[0] + w[1] * x_vals) / w[2]
        plt.plot(x_vals, y_vals, '--k')

    plt.title("Perceptr√≥n resolviendo la compuerta AND")
    plt.xlabel("Entrada 1")
    plt.ylabel("Entrada 2")
    plt.grid(True)
    plt.xlim(-0.2, 1.2)
    plt.ylim(-0.2, 1.2)
    plt.show()

plot_decision_boundary(X, y, model)
```

Aqu√≠ vemos un perceptr√≥n entrenado con la compuerta AND. Los puntos azules representan salidas falsas, y el punto rojo una salida verdadera.
 El perceptr√≥n encuentra una l√≠nea que separa correctamente las salidas. Esto es posible porque AND es una funci√≥n linealmente separable.

## Ejemplo simple con XOR

```python
# Paso 1: Importar librer√≠as
import numpy as np
import matplotlib.pyplot as plt

# Paso 2: Datos para la compuerta XOR
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 1, 1, 0])  # Salidas para XOR

# Paso 3: Funci√≥n de activaci√≥n (escal√≥n)
def step(x):
    return np.where(x >= 0, 1, 0)

# Paso 4: Perceptr√≥n simple
class SimplePerceptron:
    def __init__(self, lr=0.1, epochs=10):
        self.lr = lr
        self.epochs = epochs
        self.weights = None

    def fit(self, X, y):
        X_bias = np.hstack((np.ones((X.shape[0], 1)), X))  # Agrega columna de 1's (bias)
        self.weights = np.zeros(X_bias.shape[1])
        for _ in range(self.epochs):
            for xi, target in zip(X_bias, y):
                prediction = step(np.dot(xi, self.weights))
                self.weights += self.lr * (target - prediction) * xi

    def predict(self, X):
        X_bias = np.hstack((np.ones((X.shape[0], 1)), X))
        return step(np.dot(X_bias, self.weights))

# Paso 5: Entrenar el modelo
model = SimplePerceptron()
model.fit(X, y)
y_pred = model.predict(X)

print("Predicciones:", y_pred)

# Paso 6: Graficar frontera de decisi√≥n
def plot_decision_boundary(X, y, model):
    plt.figure(figsize=(6, 6))
    for i, point in enumerate(X):
        color = 'red' if y[i] == 1 else 'blue'
        plt.scatter(point[0], point[1], c=color, s=100, edgecolors='k')

    x_vals = np.array([-0.1, 1.1])
    w = model.weights
    if w[2] != 0:
        y_vals = -(w[0] + w[1]*x_vals) / w[2]
        plt.plot(x_vals, y_vals, '--k')

    plt.title("Perceptr√≥n intentando resolver XOR")
    plt.xlabel("Entrada 1")
    plt.ylabel("Entrada 2")
    plt.grid(True)
    plt.xlim(-0.2, 1.2)
    plt.ylim(-0.2, 1.2)
    plt.show()

plot_decision_boundary(X, y, model)
```

Aqu√≠ vemos c√≥mo el perceptr√≥n falla con XOR. Aunque intent√≥ aprender, no hay forma de separar los puntos rojos y azules con una sola l√≠nea. Esto nos lleva a una lecci√≥n muy importante: el perceptr√≥n no puede resolver problemas que no son linealmente separables. Para eso, necesitamos redes neuronales con al menos una capa oculta.
